### Задание 1: Работа с массивами NumPy
Цель: Проверить знание методов NumPy для анализа данных.

Задание:
- Создайте массив случайных чисел (50 элементов) из диапазона [10, 100].

 Найдите:
- Среднее арифметическое массива.
- Медиану.
- Стандартное отклонение.
- Создайте новый массив, где каждое число изначального массива делится на 3. Отфильтруйте числа, которые делятся без остатка.
  

```python
import numpy as np

# Создаем массив случайных чисел (50 элементов) из диапазона [10, 100]
random_array = np.random.randint(10, 101, size=50)
print("Исходный массив:", random_array)

# Находим среднее арифметическое
mean = np.mean(random_array)
print("Среднее арифметическое:", mean)

# Находим медиану
median = np.median(random_array)
print("Медиана:", median)

# Находим стандартное отклонение массива
std_dev = np.std(random_array)
print("Стандартное отклонение:", std_dev)

# Создаем новый массив, где каждое число делится на 3
divided_3 = random_array / 3
print("Массив, где каждое число разделено на 3:", divided_3)

# Фильтруем числа, которые делятся на 3 без остатка
filtered_3 = random_array[random_array % 3 == 0]
print("Числа, которые делятся на 3 без остатка:", filtered_3)
```

### Задание 2: Анализ данных с помощью Pandas
Цель: Проверить навыки анализа и обработки данных.

Данные: Используйте предоставленный CSV-файл ("sales_data_large.csv"), содержащий данные о продажах с полями:
- date (дата продаж),
- region (регион),
- product (продукт),
- sales (объем продаж).
  
Задание:
- Загрузите данные из файла.
- Определите:
- Общее количество записей.
- Какие регионы и продукты присутствуют в данных.
- Подсчитайте:
- Общие продажи по регионам.
- Средний объем продаж для каждого продукта.
- Постройте сводную таблицу, показывающую средние продажи по регионам и продуктам.

```python
import pandas as pd

# Загружаем данные из файла
df = pd.read_csv('sales_data_large.csv')

# Общее количество записей
total_rec = df.shape[0]
print("Общее количество записей:", total_rec)

# Уникальные регионы
reg = df['region'].unique()
print("Регионы в данных:", reg)

# Уникальные продукты
prod = df['product'].unique()
print("Продукты в данных:", prod)

# Общие продажи по регионам
reg_sales = df.groupby('region')['sales'].sum()
print("\nОбщие продажи по регионам:")
print(reg_sales)

# Средний объем продаж для каждого продукта
avg_prod_sales = df.groupby('product')['sales'].mean()
print("\nСредний объем продаж для каждого продукта:")
print(avg_prod_sales)

# Сводная таблица: средние продажи по регионам и продуктам
pivot_table = df.pivot_table(values='sales', index='region', columns='product', aggfunc='mean')
print("\nСводная таблица (средние продажи по регионам и продуктам):")
print(pivot_table)
```

### Задание 3: Очистка данных и визуализация
Цель: Проверить умение находить и исправлять ошибки в данных, а также визуализировать их.

Данные: Используйте предоставленный CSV-файл ("sales_cleaning_large.csv"), содержащий данные о продажах с возможными пропущенными и некорректными значениями. Поля:
- date (дата продаж),
- sales (объем продаж),
- region (регион).
  
Задание:
- Найдите и устраните ошибки в данных:
- Заполните пропущенные даты.
- Удалите отрицательные значения в столбце sales или замените их на медиану столбца.
- Заполните пропущенные значения в sales средним значением.
- Постройте график продаж по датам. Убедитесь, что даты упорядочены.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загружаем данные из файла
df = pd.read_csv('sales_cleaning_large.csv')
# Преобразуем столбец 'date' в формат datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Создаем новый столбец, преобразуем даты в числовой формат (количество дней с начала)
df['date_n'] = (df['date'] - df['date'].min()).dt.days
# Интерполируем пропущенные значения в числовом формате
df['date_n'] = df['date_n'].interpolate()
# Заполняем пропущенные даты
df.loc[df['date'].isna(), 'date'] = df['date'].min() + \
    pd.to_timedelta(df.loc[df['date'].isna(), 'date_n'], unit='D')
# Удаляем вспомогательный столбец
df = df.drop(columns=['date_n'])

# Вычисляем медиану столбца
median_s = df['sales'].median()
# Заменяем отрицательные значения на медиану
df.loc[df['sales'] < 0, 'sales'] = median_s
# Вычисляем среднее значение столбца 'sales'
mean_s = df['sales'].mean()
# Заполняем пропущенные значения средним значением
df['sales'] = df['sales'].fillna(mean_s)

# Группируем продажи по дате
df_grouped = df.groupby('date')['sales'].sum().reset_index()

# Строим график
plt.figure(figsize=(10, 6))
plt.plot(df_grouped['date'], df_grouped['sales'], marker='o', color='b', linestyle='-', label='Продажи')
plt.title('Продажи по датам')
plt.xlabel('Дата')
plt.ylabel('Продажи')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

### Задание 4: Получение данных из API
Цель: Проверить способность работать с веб-данными.

Задание:
- Используйте открытое API (например, https://jsonplaceholder.typicode.com/users).
- Загрузите данные и создайте DataFrame с полями: id, name, email, company_name.
- Выведите таблицу с пользователями, работающими в компаниях, название которых содержит слово "Group".

```python
import requests
import pandas as pd

# Загружаем данные с открытого API
url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url)
# Сохраняем данные в users
users = response.json()

# Создаем DataFrame с нужными полями
df = pd.DataFrame(users, columns=["id", "name", "email"])
# Поле 'company_name' напрямую отсутствует в JSON-объектах,
# оно внутри поля company как company["name"], добавим его отдельно
df["company_name"] = [user["company"]["name"] for user in users]
    
# Фильтруем пользователей, работающих в компаниях с названием, содержащим слово "Group"
filtered_df = df[df["company_name"].str.contains("Group", case=False, na=False)]
    
# Выводим таблицу
print("Пользователи из компаний с названием, содержащим 'Group':")
print(filtered_df)
```
